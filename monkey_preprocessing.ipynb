```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import pandas as pd
from keras.utils import to_categorical
import keras
%matplotlib inline
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.2, random_state = 101)
y_train = to_categorical(y_train, num_classes = 10)
y_valid = to_categorical(y_valid, num_classes = 10)
```

```python
import keras
from keras.models import Sequential,Input,Model
from keras.layers import Conv2D, MaxPooling2D, MaxPooling1D,Concatenate, GlobalAveragePooling2D, Dense, Dropout, Flatten, Input, LSTM, TimeDistributed
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU 
from keras.applications.inception_v3 import InceptionV3
input_tensor = Input(shape=(299,299,3))

Block1_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(input_tensor)
Block1_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block1_C1)
Block1_Out = keras.layers.concatenate([Block1_C1,Block1_R1])

Block2_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(input_tensor)
Block2_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block2_C1)
Block2_Out = keras.layers.concatenate([Block2_C1,Block2_R1])

Block3_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block1_Out)
Block3_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block3_C1)
Block3_Out = keras.layers.concatenate([Block3_C1,Block3_R1])

Block4_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block2_Out)
Block4_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block4_C1)
Block4_Out = keras.layers.concatenate([Block4_C1,Block4_R1])

Block5_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block3_Out)
Block5_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block5_C1)
Block5_Out = keras.layers.concatenate([Block5_C1,Block5_R1])

Block6_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block4_Out)
Block6_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block6_C1)
Block6_Out = keras.layers.concatenate([Block6_C1,Block6_R1])

Block7_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block5_Out)
Block7_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block7_C1)
Block7_Out = keras.layers.concatenate([Block7_C1,Block7_R1])

Block8_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block6_Out)
Block8_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block8_C1)
Block8_Out = keras.layers.concatenate([Block8_C1,Block8_R1])

Block9_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block7_Out)
Block9_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same',  activation='relu')(Block9_C1)
Block9_Out = keras.layers.concatenate([Block9_C1,Block9_R1])

Block10_C1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu',strides=(2,2))(Block8_Out)
Block10_R1 = Conv2D(filters=16, kernel_size=(5,5), padding='same', activation='relu')(Block10_C1)
Block10_Out = keras.layers.concatenate([Block10_C1,Block10_R1])

Concat_1 = keras.layers.concatenate([Block9_Out,Block10_Out])
flat_1 = Flatten()(Concat_1)
drop_out = Dropout(0.75)(flat_1)
pred = Dense(10,activation='softmax')(drop_out)


model = Model([input_tensor],pred)
model.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(x_train,y_train,batch_size = 32, epochs = 150, verbose=1,  validation_data=(x_valid, y_valid))
```
